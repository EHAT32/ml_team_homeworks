````python
from ultralytics import YOLO, SETTINGS
import os
from PIL import Image
````

Составим конфигурационный файл

````python
yaml_content = """
path: D:/python/ml_team_homeworks/license-plate-removal/datasets/generated_dataset
train: images/train
val: images/val
test: images/test
names:
    0: plate
"""
with open(f'datasets/generated_dataset/config.yaml', 'w') as f:
    f.write(yaml_content.strip())

model = YOLO("yolo11n-seg.pt")
model.train(data=f"datasets/generated_dataset/config.yaml", epochs=100)
````

````python
model = YOLO("runs/segment/train/weights/best.pt")
````
